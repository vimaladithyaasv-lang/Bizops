```python
import pandas as pd
import numpy as np

# Load the Excel file (Replace 'input.xlsx' with your actual file name)
df = pd.read_excel("C:\\Users\\s1930\\Desktop\\biz\\bizops-hackathon\\updated_file1_fixed.xlsx")

# Function to convert price to numeric
def convert_price(price):
    
    if pd.isna(price):
        return None
    price = price.replace("₹", "").replace(" ", "").replace(",", "")
    if "k" in price:
        return float(price.replace("k", "")) * 1000  # Convert 'k' to actual number
    elif "L" in price:
        return float(price.replace("L", "")) * 1e5  # Convert 'L' to lakh
    elif "Cr" in price:
        return float(price.replace("Cr", "")) * 1e7  # Convert 'Cr' to crore
    return float(price)

# Convert 'Price' column to numeric values
df['Price'] = df['Price'].apply(convert_price)

# Convert 'Size' column to numeric (remove 'Sq.Ft.')

# Group by 'Property Type', 'Size', 'Location', and 'Zone Name', then calculate the median price
grouped_df = df.groupby(['Property Type', 'Size', 'Location', 'Zone Name'], as_index=False)['Price'].median()

# Rename column for clarity
grouped_df.rename(columns={'Price': 'Median Price'}, inplace=True)

grouped_df['Size'] = grouped_df['Size'].astype(int).astype(str) + " Sq.Ft."

# Save to a new Excel file
grouped_df.to_excel("output2.xlsx", index=False)

print("File saved successfully as output.xlsx!")
```
